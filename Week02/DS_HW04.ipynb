```python
# Find tweets where 'retweet_count': {"$gt": 0}.
# What is their count?
# Is there anything about them that stands out as retweetable?
```

```python
import pandas as pd
from pymongo import MongoClient
```

```python
c = MongoClient()
```

```python
db = c.twitter
```

```python
popular_tweets = db.tweets.find({'retweet_count': {"$gte": 0}})
popular_tweets.count()
```

```python
popular_tweets = pd.DataFrame(list(db.tweets.find({'retweet_count': {"$gte": 0}})))
popular_tweets.head()
```
